# PRÁCTICA FINAL

En este primer notebook vamos a analizar las ventas diarias de 10 tiendas en 3 zonas diferentes y haremos las tranformaciones necesarias para poder realizar modelos de predicción.

```python
# Importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargamos los datos
ventas = pd.read_csv('data/raw/datos.csv', sep=',')
```

```python
# Transformamos la columna Date a tipo fecha y la usamos como índice
ventas['Date'] = pd.to_datetime(ventas['Date'])
ventas = ventas.sort_values(by=['Store', 'Date'])
ventas.set_index('Date', inplace=True)
```

```python
ventas = ventas[['Store', 'Sales']]
```

Viendo los datos vemos como del 1 julio a diciembre no hay datos de la tienda 3 en la zona b del año 2014. Por lo que vamos a rellenarlos con los datos del año anterior:

```python
fechas = ventas.loc[ventas['Store'] == 'T3b']['2013-07-01':'2013-12-31'].reset_index()
vector_fechas = pd.date_range(start='2014-07-01', end='2014-12-31')
fechas['Date'] = vector_fechas
fechas = fechas.set_index('Date')
ventas = pd.concat([ventas, fechas])
ventas = ventas.sort_values(by=['Store', 'Date'])
```

```python
ventas.loc[ventas['Store'] == 'T3b']['Store'].value_counts()
```

### Análisis de datos

```python
# ver el maximo de ventas de cada columna
ventas.groupby('Store')['Sales'].max()
```

La tienda T2c tiene el mayor volumen de ventas, 32221.

En la tienda que menos días se ha comprado es la 1 de la zona c.

Más adelante transformaremos los 0 para los modelos.

```python
# ver el maximo de ventas de cada columna
ventas['Store'].value_counts()
```

Vemos que la tienda T3b tiene menos datos por lo que lo modificaremos.

## Gráficos de las ventas diarias de las 10 tiendas

```python
# Agrupamos el dataframe por tienda y fecha y calculamos la suma de ventas por día
df_ventas = ventas.groupby(['Store', 'Date']).sum().reset_index()
plt.figure(figsize=(10,10))

for tienda in df_ventas['Store'].unique():
  ventas_tienda = df_ventas[df_ventas['Store'] == tienda]
  plt.plot(ventas_tienda['Date'], ventas_tienda['Sales'], label=tienda)

plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.show()
```

En este gráfico no se pueden diferenciar muy bien las tiendas y sus ventas, así que lo haremos tienda por tienda.

```python
# Agrupamos el dataframe por tienda y fecha y calculamos la suma de ventas por día
df_ventas = ventas.groupby(['Store', 'Date'])['Sales'].sum().reset_index()

# Obtenemos el número de tiendas
n_tiendas = len(df_ventas['Store'].unique())

# Creamos una figura con tantos subplots como tiendas tengamos
fig, axs = plt.subplots(n_tiendas, figsize=(10, 5*n_tiendas))

# Iteramos sobre las tiendas y graficamos las ventas por día en cada subplot
for i, tienda in enumerate(df_ventas['Store'].unique()):
  ventas_tienda = df_ventas[df_ventas['Store'] == tienda]
  axs[i].plot(ventas_tienda['Date'], ventas_tienda['Sales'], label=tienda)
  axs[i].set_xlabel('Fecha')
  axs[i].set_ylabel('Ventas')
  axs[i].legend()
  
```

Las tiendas T1b, T1c, T2b y T2c estuvieron cerradas durante algunos días del periodo. Más adelante lo modificaremos para una predicción más precisa.

Como podemos ver, hay muchos días en los que no se compra o por que las tiendas estaban cerradas o por que no se vendió nada.

## Gráficos de las ventas diarias por zona 

En los datos existen 3 zonas, la zona 1, la zona 2 y la zona 3.

```python
# Diferenciamos las tiendas por zona
df_ventas_zona1 = df_ventas[df_ventas['Store'].str.contains('1')]
df_ventas_zona2 = df_ventas[df_ventas['Store'].str.contains('2')]
df_ventas_zona3 = df_ventas[df_ventas['Store'].str.contains('3')]
```

```python
# graficar df_ventas_zona1, df_ventas_zona2, df_ventas_zona3
plt.figure(figsize=(10,10))
plt.plot(df_ventas_zona1['Date'], df_ventas_zona1['Sales'], label='Zona 1')
plt.plot(df_ventas_zona2['Date'], df_ventas_zona2['Sales'], label='Zona 2')
plt.plot(df_ventas_zona3['Date'], df_ventas_zona3['Sales'], label='Zona 3')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.show()
```

```python
# graficar 3 graficos en una sola figura de df_ventas_zona1, df_ventas_zona2, df_ventas_zona3
fig, axs = plt.subplots(3, figsize=(10, 15))
zonas = [df_ventas_zona1, df_ventas_zona2, df_ventas_zona3]
for i, zona in enumerate(zonas):
    axs[i].plot(zona['Date'], zona['Sales'], label='Zona {}'.format(i+1))
    axs[i].set_xlabel('Fecha')
    axs[i].set_ylabel('Ventas')
    axs[i].legend()
plt.show()
```

```python
# media de ventas por zona
media_z1 = df_ventas_zona1['Sales'].mean()
media_z2 = df_ventas_zona2['Sales'].mean()
media_z3 = df_ventas_zona3['Sales'].mean()
print('Media de ventas zona 1: {}'.format(media_z1))
print('Media de ventas zona 2: {}'.format(media_z2))   
print('Media de ventas zona 3: {}'.format(media_z3))
```

A partir de los gráficos podemos ver que la zona 3 es la que más ventas tiene, seguida de la zona 1 y la zona 2. Fijandonos en las medias la zona 2 es la que mas ventas tiene, pero como podemos ver en los gráficos, se debe ha que hay un día en el que se han vendido más de 30000 productos, por lo que la media se ve afectada por ese día.

## Transformación de datos

Para eliminar los 0 y poder hacer un mejor análisis he dicidido sustituirlos por la media móvil de orden 3.

```python
df_list = []

for tienda in ventas['Store'].unique():
    df = ventas.loc[ventas['Store'] == tienda]
    df['Sales'][0] = df['Sales'][1:4].mean()
    r = df.rolling(3).mean().shift(1)
    df = df.replace(0, np.nan)
    df = df.fillna(r)
    df_list.append(df)
    
#  ventas_df['Sales'][0] = ventas_df['Sales'][1:4].mean()


ventas_df = pd.concat(df_list)

for tienda in ventas_df['Store'].unique():
    x = ventas_df.loc[(ventas_df['Store'] == tienda) & (ventas_df['Sales'] == 0)]
    if len(x) > 0:
        print(tienda)
```
