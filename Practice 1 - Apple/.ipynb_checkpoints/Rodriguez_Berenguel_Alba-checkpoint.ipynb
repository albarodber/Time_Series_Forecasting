```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sktime.forecasting.base import ForecastingHorizon
from sktime.utils.plotting import plot_series
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.forecasting.ets import AutoETS
from sktime.performance_metrics.forecasting import mean_absolute_percentage_error
from sktime.performance_metrics.forecasting import MeanSquaredError
from sktime.forecasting.exp_smoothing import ExponentialSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
from sktime.transformations.series.boxcox import LogTransformer
from sktime.forecasting.arima import AutoARIMA
```

# 1. Preparación de los datos

```python
# Read data
apple_df = (pd.read_csv('../data/aapl_ok.csv', sep=';')
            .assign(**{'fecha': lambda df: pd.to_datetime(df['fecha'])})
            .assign(**{'Ingresos': lambda df: df['Ingresos'].str.replace(',', '.').astype(float)}))

# I remove the Quarterly column that indicates the quarters, since I will use the date one
apple_df = apple_df.drop(['Trimeste'], axis=1)

# Convert the column of dates into the index
apple_df = apple_df.set_index('fecha')

apple_df.head()
```

```python
#Convierto los datos en trimestrales 
apple_ts=apple_df.resample("q").last()
apple_ts.tail()
```

```python
#Formato SKTime
apple_ts_q=apple_df['Ingresos'].astype('float64').to_period('Q')
apple_ts_q.tail()
```

```python
#Muestro los datos en un grafico para ver mejor los movimientos
sns.set(rc={'figure.figsize':(11, 4)})
ax = apple_ts.plot(marker='.', linestyle='-')
ax.set_ylabel('Ingresos Apple por trimestres')
ax.set_xlabel('Trimestres')
ax.set_title('Evolución de los ingresos de apple por trimestres')
plt.show()
```

```python
#Grafico con el formato SKTime 
plot_series(apple_ts_q, labels=["Ventas"])
plt.show()
```

# 2. Análisis de los componentes de la serie temporal

En un primer vistazo al gráfico se puede pensar que los ingresos no son estacionarios ni en media ni en varianza, dados los
cambios que se producen.

Para confirmar esto, voy a descomponer la serie para ver sus componentes.

```python
decomposition = seasonal_decompose(apple_ts['Ingresos'], 
                                   model='multiplicative', 
                                   period=4)
decomposition.plot()
plt.show()
```

Después de ver esto se puede confirmar que los ingresos no son estacionarios ni en media ni en varianza y la tendencia fue
lineal hasta 2007 - 2008 cuando comenzo a ser creciente.

```python
ax = plt.gca()
sm.graphics.tsa.quarter_plot(apple_ts['Ingresos'],ax=ax)
ax.set_title('Comportamiento Estacional')
ax.set_ylabel('Ingresos')
ax.set_xlabel('Trimestres')
plt.show()
```

El componente estacional no es estacionario, los ingresos dependen del trimestre, en el segundo y tercer trimestre es cuando 
se registran menos ingresos, mientras que en el cuarto trimestre es cuando son más altos. 
Esto puede deberse a que Apple es una empresa que se dedica a la producción de equipos tecnológicos y al coincidir el cuarto trimestre con el período navideño se realizan más compras y son los regalos que más se suelen hacer.

# 3. Ventas anuales de Apple

Para tener un vistazo de los datos más general y ver la evolución de cada año, voy a calcular las ventas anuales de apple 
con los datos por trimestres que tengo.

```python
apple_anual = apple_ts.resample('Y').sum()
anualplot = apple_anual.plot(marker='.', linestyle='-')
anualplot.set_ylabel('Ingresos Anuales')
anualplot.set_xlabel('Años')
anualplot.set_title('Evolución de los ingresos anuales de Apple')
plt.show()
```

Las ventas fueron creciendo paulatinamente hasta 2007 cuando la empresa comenzó a ganar más fama y con ello aumentaron los
ingresos. Desde entonces la empresa no ha parado de crecer y aumentar las ventas de sus productos, a excepción del año 2016
que se produjo un descenso, debido a que disminuyeron las ventas de móviles al no presentar nuevos modelos, y este articulo es el que más ingresos les genera.
No obstante, se recuperaron en 2017 con la salida del Iphone 7 y mantuvieron esa tendencia creciente. 

La última caída corresponde al año 2021 y este no se puede tener en cuenta todavía, puesto que solo tenemos los datos para dos trimestres.

```python
apple_anual_crec = apple_anual.pct_change().mul(100)
apple_anual_crec.dropna(inplace=True)
crecplot = apple_anual_crec.plot(marker='.', linestyle='-')
crecplot.set_ylabel('Crecimiento anual (%)')
crecplot.set_xlabel('Años')
crecplot.set_title('Crecimiento de los ingresos anuales de Apple (%)')
plt.show()
```

Los mayores crecimientos se producen en 1990, 2005, 2010 y 2011. Estos años coinciden con el lanzamiento de algunos de sus
productos más vendidos y los años que hay disminución con períodos que no han sacado al mercado nuevos artículos.

En 2021 vuelve a haber una caída pronunciada, pero como he comentado antes, no la tengo en cuenta al no estar completo el año.

# 4. Modelo ETS

```python
#Separo los datos en train y test, para quitar los 8 ultimos trimestres
y_train, y_test = temporal_train_test_split(y = apple_ts['Ingresos'].astype('float64').to_period('Q'), test_size=8)

plot_series(y_train, y_test, labels=["y_train", "y_test"])
plt.show()
print(y_train.shape[0], y_test.shape[0])
```

```python
#Forecasting horizon (los 8 trimestres que quito)
fh = np.arange(1, 9)

#Forecasting algorithm
ko_auto_model = AutoETS(auto=True, sp=4, n_jobs=-1)

ko_auto_model.fit(y_train)

#Predicciones
ko_pred = ko_auto_model.predict(fh)

#Intervalo de predicciones al 90%
ko_pred_ints = ko_auto_model.predict_interval(fh, coverage=0.9)

#Grafico de las predicciones
plot_series(y_train, ko_pred, y_test, labels=["KO", "KO pred", "KO REAL"])
plt.show()
```

```python
#Grafico desde 2012, para ver mas detalladamente la prediccion
plot_series(y_train["2012":], ko_pred,y_test, labels=["KO", "KO pred", "KO REAL"])
plt.show()
```

```python
#Precision de las predicciones
mean = mean_absolute_percentage_error(y_test, ko_pred)
rmse = MeanSquaredError(square_root=True)

print("Error absoluto medio porcentual: ", mean)
print("Raiz del error cuadratico medio: ", rmse(y_test, ko_pred))
```

# 5. Modelo ARIMA

```python
#Al ser una serie no estacionaria en varianza, realizo la transformación logaritmica para que el modelo ARIMA sea correcto
transformer = LogTransformer()
log_apple_ts= transformer.fit_transform(apple_ts_q)
log_apple_ts.tail()
```

```python
# Graficos con datos logaritmicos
logplot = log_apple_ts.plot(marker='o', linestyle='-')
logplot.set_ylabel('Ingresos Apple')
logplot.set_xlabel('Trimestres')
logplot.set_title('Ingresos Apple: Transformación LOG')
plt.show()
```

```python
fig, logplot2 =plot_series(log_apple_ts, labels=["Ventas"])
logplot2.set_ylabel('Ingresos Apple')
logplot2.set_xlabel('Trimestres')
logplot2.set_title('Ingresos Apple: Transformación LOG')
plt.show()
```

```python
#Separo los datos en train y test
y_train_q, y_test_q = temporal_train_test_split(y = apple_ts_q, test_size=8)

#Train y test de la serie logaritmica
log_y_train, log_y_test = temporal_train_test_split(y = log_apple_ts, test_size=8)

#Grafico
plot_series(y_train_q, y_test_q, labels=["Train", "Test"])
plt.show()

print(y_train_q.shape[0], y_test_q.shape[0])
```

```python
# Forecast Horizon
fh = np.arange(1, 9)

#Forecasting algorithm
forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_y_train)

#Predicciones
log_y_pred = forecaster.predict(fh)

#Grafico de las predicciones
plot_series(y_train_q, np.exp(log_y_pred),y_test_q, labels=["KO", "KO pred", "KO REAL"])
plt.show()
```

```python
#Grafico desde 2012
plot_series(y_train_q["2012":], np.exp(log_y_pred),y_test_q, labels=["KO", "KO pred", "KO REAL"])
plt.show()
```

```python
#Precision de las predicciones
mean_log = mean_absolute_percentage_error(y_test, np.exp(log_y_pred))
rmse_log = MeanSquaredError(square_root=True)

print("Error absoluto medio porcentual: ", mean_log)
print("Raiz del error cuadratico medio: ", rmse_log(y_test, np.exp(log_y_pred)))
```

# 6. Comparación de los modelos y cálculo de las predicciones 2021 - 2022

Una vez hechos ambos modelos, voy a comparar sus resultados con los datos reales, para comprobar cuál tiene menor error en sus 
predicciones y usarlo para pronosticar los ingresos de los trimestres restante de 2021 y 2022.

```python
plot_series(y_train["2012":], y_test, ko_pred, np.exp(log_y_pred), labels=["KO", "KO REAL", "KO ETS", "KO ARIMA"])
plt.show()
```

En el gráfico se puede ver los ingresos reales y los calculados con el modelo ETS y modelo ARIMA. A simple vista se ve que los
del modelo ARIMA están un poco por encima de los del modelo ETS, más cercanos a los datos reales.

Para verificar esto comparo las metricas de precisión que he calculado antes:

```python
if mean < mean_log:
    print ("La mean del modelo ETS es: ", round(mean,4), " y es inferior a la del modelo ARIMA")
else:
    print("La mean del modelo ARIMA es:", round(mean_log,4), "y es inferior a la del modelo ETS")
    
if rmse(y_test, ko_pred[0:8]) < rmse_log(y_test, np.exp(log_y_pred)): 
    print ("El rmse del modelo ETS es: ", round(rmse(y_test, ko_pred),4), " y es inferior a la del modelo ARIMA")
else:
    print("El rmse del modelo ARIMA es:", round(rmse_log(y_test, np.exp(log_y_pred)),4), "y es inferior a la del modelo ETS")
```

Vista la grafica y las precisiones de los modelos, voy a escoger el modelo ARIMA para realizar las predicciones ya que es más 
preciso y nos devuelve mejores datos. Tiene un error absoluto medio porcentual del 13,97% frente al 16,27% del modelo ETS, lo 
que quiere decir que se desvía menos de los datos reales.

Para ello, ahora sí voy a incluir los últimos 8 trimestres, puesto que ya he comparado los modelos.

```python
# Forecasting horizon
fh = np.arange(1, 7) 

#Forecasting algorithm
forecaster = AutoARIMA(sp=4,suppress_warnings=True)
forecaster.fit(log_apple_ts)

#Predicciones
apple_2122_pred = forecaster.predict(fh)
plot_series(apple_ts_q, np.exp(apple_2122_pred), labels=["KO", "KO pred"])
plt.show()
```

```python
plot_series(apple_ts_q["2015":], np.exp(apple_2122_pred), labels=["KO", "KO pred"])
plt.show()
```

En consecuencia, las ventas anuales de 2021 y 2022 serían:

```python
apple_anual_pred = pd.concat([(apple_ts_q[-2:]), (np.exp(apple_2122_pred))]).resample('Y').sum()

print("Las ventas acumuladas en 2021 serán:", round(apple_anual_pred[0], 4), "$")
print("Las ventas acumuladas en 2022 serán:", round(apple_anual_pred[1], 4), "$")
```

# 7. ¿Efecto COVID?

La pandemia por el COVID-19 comenzó en marzo de 2020, finalizando el primer trimestre del año. Para frenar esta enfermedad
se introdujeron medidas en todos los países, que afectaron negativamente a muchas empresas a nivel mundial.

Para saber como afectó el COVID a Apple voy a analizar sus ventas desde un período más cercano al comienzo de la pandemia y 
cómo evolucionaron desde entonces, y lo voy a comparar con las predicciones hechas con el modelo ARIMA para los últimos 8 
trimestres de la serie.

```python
plot_series(apple_ts_q["2018":], np.exp(log_y_pred), labels=["KO REAL", "KO pred"])
plt.show()
```

En el gráfico se puede ver que las ventas desde que comenzó el COVID han sido superiores a la predichas por el modelo.
Esto puede deberse a la transformación que ha habido en las empresas o en los colegios para adaptar la forma de trabajar o la de
enseñar a un modelo COVID, que permitiese no parar estas actividades. Para poder hacer esto posible, la venta de ordenadores y otros equipos tecnológicos se disparó.

Voy a cuantificar cuánto ha sido el efecto COVID exactamente. Para ello, voy a fijar que el COVID dió comienzo en el segundo
trimestre de 2020 (2020Q2), puesto que cuando comenzaron a implementarse más medidas restrictivas internacionales fue en la segunda quincena de marzo, por lo que no considero óptimo incluir el trimestre entero. Y no voy a establecer ninguna fecha de finalización, ya que el final de la serie es el segundo trimestre de 2021 y en esa fecha aun seguía habiendo algunas medidas
severas.

```python
COVID_real = y_test["2020Q2":].sum()
COVID_pred = round(np.exp(log_y_pred)["2020Q2":].sum(), 2)

print("En el segundo trimestre de 2020 ingresaron", round(y_test["2020Q2":] - np.exp(log_y_pred)["2020Q2":], 2)[0], 
      "$ más de lo estimado")
print("En el tercer trimestre de 2020 ingresaron", round(y_test["2020Q2":] - np.exp(log_y_pred)["2020Q2":], 2)[1],
     "$ más de lo estimado")
print("En el cuarto trimestre de 2020 ingresaron", round(y_test["2020Q2":] - np.exp(log_y_pred)["2020Q2":], 2)[2],
     "$ más de lo estimado")
print("En el primer trimestre de 2021 ingresaron", round(y_test["2020Q2":] - np.exp(log_y_pred)["2020Q2":], 2)[3],
     "$ más de lo estimado")
print("En el segundo trimestre de 2021 ingresaron", round(y_test["2020Q2":] - np.exp(log_y_pred)["2020Q2":], 2)[4],
     "$ más de lo estimado")

if COVID_real < COVID_pred:
    print("El efecto COVID ha sido negativo. Han ingresado", round(COVID_pred - COVID_real, 2), "$ menos de lo estimado")
else:
    print("El efecto COVID ha sido positivo. Han ingresado", round(COVID_real - COVID_pred, 2), "$ más de lo estimado")
```
