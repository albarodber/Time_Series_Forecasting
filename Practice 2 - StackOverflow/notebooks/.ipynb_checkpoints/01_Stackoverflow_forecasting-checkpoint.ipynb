The objective of this work is to analyze the evolution of questions on MATLAB in StackOverFlow and build a model to predict the 12 months out of the sample. For this I am going to use the ETS, ARIMA, 4Theta and TBATS models and I will decide which one gives me more precise results.

Steps to perform:

1. Data preparation
2. Data analysis
3. ETS model
4. ARIMA model
5. 4Theta Model
6. TBATS model
7. Comparison of models and calculation of predictions

```python
# Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sktime.utils.plotting import plot_series
from sktime.transformations.series.boxcox import LogTransformer
import statsmodels.api as sm
from statsmodels.tsa.seasonal import STL
from sktime.forecasting.model_selection import temporal_train_test_split
from darts import TimeSeries
from darts.models import ExponentialSmoothing
from darts.models import TBATS
from darts.metrics import mape
from darts.models import FourTheta
from darts.models import ARIMA
from sktime.forecasting.base import ForecastingHorizon
from darts.utils.utils import SeasonalityMode, TrendMode, ModelMode
import warnings
warnings.filterwarnings('ignore')
```

# 1. Data preparation

```python
# Read data
stack_df = pd.read_csv('../data/StackOverflow.csv', sep=',')

# Filter the columns months and matlab       
stack_df = stack_df.iloc[:, [0, 8]]

#The month column was in the format 09-Jan (year and month), so it gave me problems to transform it to date format or
#do operations with it. I replace it with a range of dates between 01-31-2009 and 12-31-2019, which is the period
#that has the data frame, 09-Jan to 19-Dec.
stack_df['month'] = pd.date_range(start='2009-01-31', end='2019-12-31', freq='m')

# Convert the month column into the index
stack_df = stack_df.set_index('month')

stack_df.head()
```

```python
# Convert the data to monthly
stack_ts=stack_df.resample("m").last()
stack_ts.tail()
```

```python
# SKTime format
stack_ts_m=stack_df['matlab'].astype('float64').to_period('m').sort_index()
stack_ts_m.tail()
```

```python
# I convert the data to quarterly
stack_q=stack_df.resample("q").last()
stack_q.tail()
```

```python
# I perform the logarithmic transformation because it is necessary to check the stationarity in variance
transformer = LogTransformer()
log_stack_ts= transformer.fit_transform(stack_ts_m)
log_stack_ts.tail()
```

### Display of data in different periods

```python
sns.set(rc={'figure.figsize':(10, 3)})
ax = stack_q.plot(marker='.', linestyle='-')
ax.set_ylabel('Preguntas')
ax.set_xlabel('Trimestres')
ax.set_title('Evolución trimestral de las preguntas sobre MATLAB')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10, 3)})
ax = stack_df.plot(marker='.', linestyle='-')
ax.set_ylabel('Preguntas')
ax.set_xlabel('Meses')
ax.set_title('Evolución mensual de las preguntas sobre MATLAB')
plt.show()
```

```python
plot_sktime = plot_series(stack_ts_m, labels=["MATLAB"])
plt.xlabel('Trimestres')
plt.ylabel('Preguntas')
plt.title('Evolución de las preguntas sobre MATLAB (SKTIME)')
plt.show()
```

# 2. Data analysis

### Stationarity and seasonality

```python
plot_series(stack_ts_m, labels=["Preguntas MATLAB"])
plt.ylabel('Preguntas')
plt.xlabel('Fecha')
plt.title('Preguntas MATLAB')
plt.show()
```

```python
plot_series(log_stack_ts, labels=["Log Preguntas MATLAB"])
plt.ylabel('Preguntas')
plt.xlabel('Fecha')
plt.title('Preguntas MATLAB')
plt.show()
```

With these two graphs it can be affirmed that the time series is not stationary neither in mean nor in variance, since neither of them remains constant over time.

```python
ax = plt.gca()
sm.graphics.tsa.month_plot(stack_ts['matlab'],ax=ax)
ax.set_title('Comportamiento Estacional')
ax.set_ylabel('Preguntas')
ax.set_xlabel('Meses')
plt.show()
```

The greatest number of questions is made in the period from March to May, with the highest being April. This may be because it is widely used in academic and business fields and this period coincides with the end of the first quarter of the year and the beginning of the second, when many companies have to carry out new analyses; the students are close to the end of the course, so there will be more users who start using it, and therefore, doubts will be generated among them.

The month in which there are fewer questions is August, which coincides with the holiday period.

Finally, I'm going to decompose the series to see its components.

```python
res = STL(stack_df['matlab']).fit()
res.plot()
plt.show()
```

As a conclusion to the graphs, the series is not stationary in mean or variance, but it is seasonal. Its trend was increasing until 2014 when it began to decrease.

### Annual questions about MATLAB

```python
stack_year=stack_ts_m.resample('A').sum()

anualplot = stack_year.plot(marker='.', linestyle='-')
anualplot.set_ylabel('Preguntas')
anualplot.set_xlabel('Años')
anualplot.set_title('Evolución anual de las preguntas sobre MATLAB')
plt.show()
```

The data goes from 2009 to 2019, you can see a growth in questions until 2014, when they begin to decline.

### Annual growth of MATLAB questions

```python
stack_year_crec = stack_year.pct_change()
stack_year_crec.dropna(inplace=True)

crecplot = stack_year_crec.plot(marker='.', linestyle='-')
crecplot.set_ylabel('Crecimiento anual')
crecplot.set_xlabel('Años')
crecplot.set_title('Crecimiento anual de las preguntas sobre MATLAB')
plt.show()
```

Regarding growth, the largest occurred in 2010 and 2011, from 2015 they begin to be negative since the number of questions falls.

# 3. ETS Model

```python
stack_serie = TimeSeries.from_dataframe(stack_df)

# I separate the data into train and test, so that the last 6 months of the sample are tests
y_train_ets, y_test_ets = stack_serie.split_before(pd.Timestamp("20190801"))

y_train_ets.plot(label="training")
y_test_ets.plot(label="test")


print(len(y_train_ets), len(y_test_ets))
```

```python
# Model
modelo_ets = ExponentialSmoothing(trend=ModelMode.MULTIPLICATIVE, seasonal=SeasonalityMode.ADDITIVE, seasonal_periods=12)
modelo_ets.fit(y_train_ets)

# Predictions
pred_ets = modelo_ets.predict(len(y_test_ets))

y_train_ets.plot(label="train")
y_test_ets.plot(label="true")
pred_ets.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_ets, mape(y_test_ets, pred_ets)))
```

# 4. ARIMA Modl

```python
y_train_ar, y_test_ar = stack_serie.split_before(pd.Timestamp("20190801"))

y_train_ar.plot(label="training")
y_test_ar.plot(label="test")


print(len(y_train_ar), len(y_test_ar))
```

```python
# Model
modelo_ar = ARIMA(p=0, d=0, q=1, seasonal_order=(1, 0, 0, 12))
modelo_ar.fit(y_train_ar)

# Predictions
pred_ar = modelo_ar.predict(len(y_test_ar))

y_train_ar.plot(label="train")
y_test_ar.plot(label="true")
pred_ar.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_ar, mape(y_test_ar, pred_ar)))
```

# 5. 4Theta Model

```python
# Divided into train and test
y_train_theta, y_test_theta = stack_serie.split_before(pd.Timestamp("20190801"))

# Divided into validation
y_train_theta2, y_val_theta = y_train_theta.split_before(pd.Timestamp("20180801"))

y_train_theta2.plot(label="training")
y_val_theta.plot(label="validation")
y_test_theta.plot(label="test")


print(len(y_train_theta2), len(y_val_theta), len(y_test_theta))
```

```python
# I make a first 4Theta model for the validation part, without specific parameters
modelo_theta =FourTheta(seasonality_period=12)
modelo_theta.fit(y_train_theta2)

forecast_theta = modelo_theta.predict(len(y_val_theta))

print("model {} obtains MAPE: {:.2f}%".format(modelo_theta, mape(y_val_theta, forecast_theta)))

y_train_theta2.plot(label="train")
y_val_theta.plot(label="true")
forecast_theta.plot(label="prediction")
```

The returned model has too high an error, I'm going to use gridsearch() to find the best model:

```python
theta_grid = {
    'theta':[-4,-3,-2,-1,1,2,3,4],
    'model_mode': [ModelMode.ADDITIVE,ModelMode.MULTIPLICATIVE],
    'season_mode': [SeasonalityMode.MULTIPLICATIVE,SeasonalityMode.ADDITIVE],
    'trend_mode': [TrendMode.EXPONENTIAL,TrendMode.LINEAR]
}

best_grid_model=FourTheta.gridsearch(parameters=theta_grid,
                                series=y_train_theta2,
                                forecast_horizon=12,
                                start=0.5,
                                last_points_only=False,
                                metric=mape,
                                reduction=np.mean,
                                verbose=False,
                                n_jobs=-1)

print(best_grid_model)
```

I set the obtained parameters: *<ModelMode.ADDITIVE: 'additive'>, 'season_mode': <SeasonalityMode.ADDITIVE: 'additive'>, 'trend_mode': <TrendMode.EXPONENTIAL: 'exponential'>*

I apply it to the validation set to make a first check of the improvement of the model:

```python
modelo_theta2=FourTheta(theta=best_grid_model[1]['theta'],
                 model_mode=ModelMode.ADDITIVE,
                 season_mode=SeasonalityMode.ADDITIVE,
                 trend_mode=TrendMode.EXPONENTIAL
                 )

modelo_theta2.fit(y_train_theta2)
pred_theta = modelo_theta2.predict(len(y_val_theta))

y_train_theta2.plot(label="train")
y_val_theta.plot(label="true")
pred_theta.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_theta2, mape(y_val_theta, pred_theta)))
```

The model has improved considerably, I apply it to the test to confirm it:

```python
modelo_theta2.fit(y_train_theta)
pred_theta2 = modelo_theta2.predict(len(y_test_theta))

y_train_theta.plot(label="train")
y_test_theta.plot(label="true")
pred_theta2.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_theta2, mape(y_test_theta, pred_theta2)))
```

# 6. TBATS Model

```python
y_train_tb, y_test_tb = stack_serie.split_before(pd.Timestamp("20190801"))
y_train_tb2, y_val_tb = y_train_tb.split_before(pd.Timestamp("20180801"))

y_train_tb2.plot(label="training")
y_val_tb.plot(label="validation")
y_test_tb.plot(label="test")


print(len(y_train_tb2), len(y_val_tb), len(y_test_tb))
```

```python
modelo_tb =TBATS(use_box_cox=None, box_cox_bounds=(0, 1), use_trend=None, use_damped_trend=None, seasonal_periods="freq", 
                 use_arma_errors=True, show_warnings=False, multiprocessing_start_method='spawn', random_state=0)

modelo_tb.fit(y_train_tb2)

forecast_tb = modelo_tb.predict(len(y_val_tb))

y_train_tb2.plot(label="train")
y_val_tb.plot(label="true")
forecast_tb.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_tb, mape(forecast_tb, y_val_tb)))
```

The model obtained has a somewhat high MAPE, I am going to modify some of its parameters and try to obtain a better one:

```python
modelo_tb2 =TBATS(use_box_cox=True, box_cox_bounds=(0, 1), use_trend=False, use_damped_trend=False, seasonal_periods=[12.],
                  use_arma_errors=True, show_warnings=False, multiprocessing_start_method='spawn', random_state=0)

modelo_tb2.fit(y_train_tb2)

forecast_tb2 = modelo_tb2.predict(len(y_val_tb))

y_train_tb2.plot(label="train")
y_val_tb.plot(label="true")
forecast_tb2.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_tb2, mape(forecast_tb2, y_val_tb)))
```

The MAPE of the model has been reduced by 14%, finally, I am going to apply this model to the test part:

```python
modelo_tb2.fit(y_train_tb)

forecast_tb3 = modelo_tb2.predict(len(y_test_tb))

y_train_tb.plot(label="train")
y_test_tb.plot(label="true")
forecast_tb3.plot(label="prediction")

print("model {} obtains MAPE: {:.2f}%".format(modelo_tb2, mape(forecast_tb3, y_test_tb)))
```

# 7. Comparison of models and calculation of predictions

At this point I am going to compare the models built with the objective of selecting the one that returns the most accurate results and being able to calculate the predictions for 2020. With what was calculated in the previous sections, this is the comparison of the four models with reality ( applied to the test part):

```python
# I transform the results to df
real_df = y_test_ets.pd_dataframe()
pred_ets_df = pred_ets.pd_dataframe()
pred_ar_df = pred_ar.pd_dataframe()
pred_theta_df = pred_theta2.pd_dataframe()
pred_tb_df = forecast_tb3.pd_dataframe()

# Create a df with the results
test_comp = pd.concat([real_df, pred_ets_df, pred_ar_df, pred_theta_df, 
                       pred_tb_df], axis=1).rename(columns={0 : 'REAL', 1 : 'ETS', 2 : 'ARIMA',3 : '4THETA',4 : 'TBAST'})

test_comp.columns = ["REAL", "ETS", "ARIMA", "4THETA", "TBAST"]

test_comp
```

To check the models built more accurately, I will use cross-validation with each of the best models that I have obtained in each section and I will choose the one that best fits the original series, returning a minor error.

```python
#ETS
historical_fcast_ets = modelo_ets.historical_forecasts(
    stack_serie, start=0.4, forecast_horizon=12, verbose=True, stride=1
)

stack_serie.plot(label="data")
historical_fcast_ets.plot(label="backtest 12-months ahead forecast (Tbats)")
print("MAPE = {:.2f}%".format(mape(historical_fcast_ets, stack_serie)))
```

```python
#ARIMA
historical_fcast_ar = modelo_ar.historical_forecasts(
    stack_serie, start=0.4, forecast_horizon=12, verbose=True, stride=1
)

stack_serie.plot(label="data")
historical_fcast_ar.plot(label="backtest 12-months ahead forecast (Tbats)")
print("MAPE = {:.2f}%".format(mape(historical_fcast_ar, stack_serie)))
```

```python
#4THETA
historical_fcast_theta = modelo_theta2.historical_forecasts(
    stack_serie, start=0.4, forecast_horizon=12, verbose=True, stride=1
)

stack_serie.plot(label="data")
historical_fcast_theta.plot(label="backtest 12-months ahead forecast (Tbats)")
print("MAPE = {:.2f}%".format(mape(historical_fcast_theta, stack_serie)))
```

```python
#TBATS
historical_fcast_tbats = modelo_tb2.historical_forecasts(
    stack_serie, start=0.4, forecast_horizon=12, verbose=True, stride=1
)

stack_serie.plot(label="data")
historical_fcast_tbats.plot(label="backtest 12-months ahead forecast (Tbats)")
print("MAPE = {:.2f}%".format(mape(historical_fcast_tbats, stack_serie)))
```

Previously, when calculating the models, I also calculated the MAPE and after applying it to the test set, the best was obtained with the ARIMA model. However, with the TBATS model, I obtained the lowest MAPE but when applying it to the validation set.

With the cross-validation, I confirm that the best model is the TABTS, since it is more in line with the original series and its error is lower, so it will be the one I choose to make the 2020 predictions.

```python
# I apply the best tbast model to the series
modelo_tb2.fit(stack_serie)

pred_20 = modelo_tb2.predict(12)

stack_serie.plot(label="REAL")
pred_20.plot(label="pred")
```
